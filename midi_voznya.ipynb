```python
import pretty_midi
import pandas as pd
```

```python
# Загрузка MIDI файла
midi_data = pretty_midi.PrettyMIDI('HarryPotterPrologue.mid')
notes_by_instruments = dict()
instrument_number = 0
for instrument in midi_data.instruments:
    partiture = []
    for note in instrument.notes:
        partiture.append([note.pitch, note.start, note.end, note.velocity])
    notes_by_instruments[instrument_number] = partiture
    instrument_number += 1
notes_by_instruments
```

```python
reversed_midi = pretty_midi.PrettyMIDI()

Electric_Piano_1 = pretty_midi.Instrument(program=4)
Tremolo_Strings = pretty_midi.Instrument(program=44)
String_Ensemble_1_1 = pretty_midi.Instrument(program=48)
Pizzicato_Strings_1 = pretty_midi.Instrument(program=45)
String_Ensemble_1_2 = pretty_midi.Instrument(program=48)
Flute = pretty_midi.Instrument(program=73)
Clarinet = pretty_midi.Instrument(program=71)
Timpani = pretty_midi.Instrument(program=47)
String_Ensemble_1_3 = pretty_midi.Instrument(program=48)
Pizzicato_Strings_2 = pretty_midi.Instrument(program=45)
Pizzicato_Strings_3 = pretty_midi.Instrument(program=45)
String_Ensemble_1_4 = pretty_midi.Instrument(program=48)
French_Horn = pretty_midi.Instrument(program=60)
Oboe = pretty_midi.Instrument(program=68)
String_Ensemble_1_5 = pretty_midi.Instrument(program=48)
Pizzicato_Strings_4 = pretty_midi.Instrument(program=45)

for row in notes_by_instruments[0]:
    note = pretty_midi.Note(
        velocity=int(row[3]),
        pitch=int(140-row[0]),
        start=float(row[1]),
        end=float(row[2])
    )
    Electric_Piano_1.notes.append(note)

for row in notes_by_instruments[1]:
    note = pretty_midi.Note(
        velocity=int(row[3]),
        pitch=int(140-row[0]),
        start=float(row[1]),
        end=float(row[2])
    )
    Tremolo_Strings.notes.append(note)

for row in notes_by_instruments[2]:
    note = pretty_midi.Note(
        velocity=int(row[3]),
        pitch=int(140-row[0]),
        start=float(row[1]),
        end=float(row[2])
    )
    String_Ensemble_1_1.notes.append(note)

for row in notes_by_instruments[3]:
    note = pretty_midi.Note(
        velocity=int(row[3]),
        pitch=int(140-row[0]),
        start=float(row[1]),
        end=float(row[2])
    )
    Pizzicato_Strings_1.notes.append(note)

for row in notes_by_instruments[4]:
    note = pretty_midi.Note(
        velocity=int(row[3]),
        pitch=int(140-row[0]),
        start=float(row[1]),
        end=float(row[2])
    )
    String_Ensemble_1_2.notes.append(note)

for row in notes_by_instruments[5]:
    note = pretty_midi.Note(
        velocity=int(row[3]),
        pitch=int(140-row[0]),
        start=float(row[1]),
        end=float(row[2])
    )
    Flute.notes.append(note)

for row in notes_by_instruments[6]:
    note = pretty_midi.Note(
        velocity=int(row[3]),
        pitch=int(140-row[0]),
        start=float(row[1]),
        end=float(row[2])
    )
    Clarinet.notes.append(note)

for row in notes_by_instruments[7]:
    note = pretty_midi.Note(
        velocity=int(row[3]),
        pitch=int(140-row[0]),
        start=float(row[1]),
        end=float(row[2])
    )
    String_Ensemble_1_3.notes.append(note)
    
for row in notes_by_instruments[8]:
    note = pretty_midi.Note(
        velocity=int(row[3]),
        pitch=int(140-row[0]),
        start=float(row[1]),
        end=float(row[2])
    )
    Pizzicato_Strings_2.notes.append(note)
    
for row in notes_by_instruments[9]:
    note = pretty_midi.Note(
        velocity=int(row[3]),
        pitch=int(140-row[0]),
        start=float(row[1]),
        end=float(row[2])
    )
    Pizzicato_Strings_3.notes.append(note)
    
for row in notes_by_instruments[10]:
    note = pretty_midi.Note(
        velocity=int(row[3]),
        pitch=int(140-row[0]),
        start=float(row[1]),
        end=float(row[2])
    )
    String_Ensemble_1_4.notes.append(note)

for row in notes_by_instruments[11]:
    note = pretty_midi.Note(
        velocity=int(row[3]),
        pitch=int(140-row[0]),
        start=float(row[1]),
        end=float(row[2])
    )
    French_Horn.notes.append(note)

for row in notes_by_instruments[12]:
    note = pretty_midi.Note(
        velocity=int(row[3]),
        pitch=int(140-row[0]),
        start=float(row[1]),
        end=float(row[2])
    )
    Oboe.notes.append(note)

for row in notes_by_instruments[13]:
    note = pretty_midi.Note(
        velocity=int(row[3]),
        pitch=int(140-row[0]),
        start=float(row[1]),
        end=float(row[2])
    )
    String_Ensemble_1_5.notes.append(note)
    
for row in notes_by_instruments[14]:
    note = pretty_midi.Note(
        velocity=int(row[3]),
        pitch=int(140-row[0]),
        start=float(row[1]),
        end=float(row[2])
    )
    Pizzicato_Strings_4.notes.append(note)

    

reversed_midi.instruments.append(Electric_Piano_1)
reversed_midi.instruments.append(Tremolo_Strings)
reversed_midi.instruments.append(String_Ensemble_1_1)
reversed_midi.instruments.append(Pizzicato_Strings_1)
reversed_midi.instruments.append(String_Ensemble_1_2)
reversed_midi.instruments.append(Flute)
reversed_midi.instruments.append(Clarinet)
reversed_midi.instruments.append(String_Ensemble_1_3)
reversed_midi.instruments.append(Pizzicato_Strings_2)
reversed_midi.instruments.append(Pizzicato_Strings_3)
reversed_midi.instruments.append(String_Ensemble_1_4)
reversed_midi.instruments.append(French_Horn)
reversed_midi.instruments.append(Oboe)
reversed_midi.instruments.append(String_Ensemble_1_5)
reversed_midi.instruments.append(Pizzicato_Strings_4)

# Сохранение MIDI файла
reversed_midi.write('midi reversed.mid')
```
